```python
import yfinance as yf
import pandas as pd

pd.set_option("display.max_rows", None)

data =  yf.download("SMCI", period="5y")
data.drop(columns=["Close"], inplace=True)

data["Percent Change"] = data["Adj Close"].pct_change() * 100
data["Next Day Return"] = data["Percent Change"].shift(-1)
data.dropna(subset=["Percent Change", "Next Day Return"], inplace=True)

print(data)
```
