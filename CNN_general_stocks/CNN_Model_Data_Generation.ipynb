```python
data = CNN_general_stocks.get_features_LSTM(tickers)

sequence_length = 60
X_train = []
y_train = []

X_val = []
y_val = []

X_test = []
y_test = []
for d in data:
     features = d.copy()
     features = features.drop(columns=["Next_Day_Return"])

     train_num = (len(d) - sequence_length) * 0.6
     val_num = (len(d) - sequence_length) * 0.2
     test_num = (len(d) - sequence_length) * 0.2
     for i in range(len(d) - sequence_length):
          X_window = features.iloc[i : i + sequence_length].values

          day_return = (d.iloc[i + sequence_length]["Next_Day_Return"])
          
          if(i < train_num):
               X_train.append(X_window)
               y_train.append(day_return)
          elif (i < train_num + val_num):
               X_val.append(X_window)
               y_val.append(day_return)
          else:
               X_test.append(X_window)
               y_test.append(day_return)


X_train = np.array(X_train)
y_train = np.array(y_train)

X_val = np.array(X_val)
y_val = np.array(y_val)

X_test = np.array(X_test)
y_test = np.array(y_test)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
print(X_val.shape)
print(y_val.shape)
```
